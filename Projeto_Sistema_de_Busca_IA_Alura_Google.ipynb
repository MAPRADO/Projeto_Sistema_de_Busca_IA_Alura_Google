Vamos criar nossa função utilizando outro modelo

```python
model = "models/embedding-001"
```

```python
def embed_fn(title, text):
  return genai.embed_content(model=model,
                                content=text,
                                title=title,
                                task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

Acessar o data frame e criar uma coluna com qualquer nome (Embeddings) que vai receber o retorno da função, usando o método "apply" do pandas. Usar a função lambda (funciona como um for) para fazer o apply.

```python
df["Embeddings"] = df.apply(lambda row: embed_fn(row["Titulo"], row["Titulo"]), axis=1)
display(df)
```

Criar uma função para fazer a consulta em formato de embedding, para encontrar o arquivo.  
Usando o numpay para fazer o calculo matemático que vai calcular a menor distancia do token, para correlacionar e identificar a consulta com o documento por semântica.

```python
def gerar_e_buscar_consulta(consulta, base, model):
  embedding_da_consulta = genai.embed_content(model=model,
                                content=consulta,
                                task_type="RETRIEVAL_QUERY")["embedding"]

  produtos_escalares = np.dot(np.stack(df["Embeddings"]), embedding_da_consulta)

  indice = np.argmax(produtos_escalares)
  return df.iloc[indice]["Conteudo"]
```

Fazemos a pergunta para o modelo e chamamos a função para identificar o documento.

```python
consulta = "Um idoso com emorragia?"
trecho = gerar_e_buscar_consulta(consulta, df, model)
display(trecho)
```
